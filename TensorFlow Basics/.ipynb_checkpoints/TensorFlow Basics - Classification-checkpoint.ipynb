## TF Basics - Classification

```python
import pandas as pd
```

```python
diabetes = pd.read_csv('pima-indians-diabetes.csv')
```

```python
diabetes.head()
```

```python
diabetes.columns
```

```python
# can't normalize class due to label column
# can't normalize group column due to strings
# going to categorize age, therefore removing
cols_to_norm = ['Number_pregnant', 'Glucose_concentration', 'Blood_pressure', 'Triceps',
       'Insulin', 'BMI', 'Pedigree']
```

```python
# pandas let's you quickly normalize data
diabetes[cols_to_norm] = diabetes[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```python
diabetes.head()
```

```python
import tensorflow as tf
```

```python
diabetes.columns
```

```python
# continuous values
```

```python
num_preg = tf.feature_column.numeric_column('Number_pregnant')
```

```python
plasma_gluc = tf.feature_column.numeric_column('Glucose_concentration')
```

```python
dias_press = tf.feature_column.numeric_column('Blood_pressure')
```

```python
tricep = tf.feature_column.numeric_column('Triceps')
```

```python
insulin = tf.feature_column.numeric_column('Insulin')
```

```python
bmi = tf.feature_column.numeric_column('BMI')
```

```python
diabetes_pedigree = tf.feature_column.numeric_column('Pedigree')
```

```python
age = tf.feature_column.numeric_column('Age')
```

```python
# Categorical values can either be done with a vocabulary list or hash bucket
```

```python
assigned_group = tf.feature_column.categorical_column_with_vocabulary_list('Group', ['A', 'B', 'C', 'D'])
```

```python
# hash bucket size is the max number of groups you believe to be in the sample
#assigned_group = tf.feature_column.categorical_column_with_hash_bucket('Group', hash_bucket_size=10)
```

```python
# FEATURE ENGINEERING
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diabetes['Age'].hist(bins=20)
```

```python
# taking a continuous value and putting it into buckets
age_bucket = tf.feature_column.bucketized_column(age, boundaries=[20,30,40,50,60,70,80])
```

```python
feat_cols = [num_preg, plasma_gluc, dias_press, tricep, insulin, bmi, diabetes_pedigree, assigned_group, age_bucket]
```

```python
# TRAIN TEST SPLIT
x_data = diabetes.drop('Class', axis=1)
```

```python
x_data.head()
```

```python
labels = diabetes['Class']
labels.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, labels, test_size=0.3, random_state=101)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=10, num_epochs=1000, shuffle=True)
```

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols, n_classes=2)
```

```python
 model.train(input_fn=input_func, steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test, y=y_test, batch_size=10, num_epochs=1, shuffle=False)
```

```python
results = model.evaluate(eval_input_func)
```

```python
results
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test, batch_size=10, num_epochs=1, shuffle=False)
```

```python
predictions = model.predict(pred_input_func)
```

```python
my_pred = list(predictions)
```

```python
my_pred
```

```python
# dense neuralm network
dnn_model = tf.estimator.DNNClassifier(hidden_units=[10, 10, 10], feature_columns=feat_cols, n_classes=2)
```

```python
embedded_group_col = tf.feature_column.embedding_column(assigned_group, dimension=4)
```

```python
feat_cols = [num_preg, plasma_gluc, dias_press, tricep, insulin, bmi, diabetes_pedigree, embedded_group_col, age_bucket]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(X_train, y_train, batch_size=10, num_epochs=1000, shuffle=True)
```

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[10,20, 20, 20, 10], feature_columns=feat_cols, n_classes=2)
```

```python
dnn_model.train(input_fn=input_func, steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test,y=y_test,batch_size=10,num_epochs=1, shuffle=False)
```

```python
dnn_model.evaluate(eval_input_func)
```
